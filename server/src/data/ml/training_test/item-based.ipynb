```python
df = df.loc[:, no_movies_voted[no_movies_voted > 50].index]
```

```python
df
```

```python
df.head()
```

```python
from scipy.sparse import csr_matrix
```

```python
csr_data = csr_matrix(df.values)
df.reset_index(inplace=True)
```

```python
print(csr_data)
```

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)
knn.fit(csr_data)
```

```python
def get_recommendation(movie_name):
  movie_list = movie_df[movie_df['title'].str.contains(movie_name)]
  if len(movie_list):
    print("Movies Found")
    movie_idx = movie_list.iloc[0]['movieId']
    user_idx = df[df['movieId'] == movie_idx].index
    print(user_idx, movie_idx)
    distance, indices = knn.kneighbors(csr_data[user_idx], n_neighbors=10+1)
    print(distance, indices)
    rec_movies = sorted(list(zip(indices.squeeze().tolist(), distance.squeeze().tolist())), key=lambda x: x[1])[:0: -1]
    print(rec_movies)
    
    for val in rec_movies:
      movie_idx = df.iloc[val[0]]['movieId']
      print(movie_df[movie_df['movieId'] == movie_idx])
      
  else:
    print("Movie not Found")
```

```python
get_recommendation('Iron Man 2')
```
